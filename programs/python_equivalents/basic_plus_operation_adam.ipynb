```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers

# Generate training data
# In this example, we'll generate random pairs of numbers for training.
# You can replace this with your own dataset if needed.
num_samples = 10000
# Generate random pairs of numbers between 0 and 100
X = np.random.randint(0, 101, size = (num_samples, 2))
# Calculate the sum of each pair
y = np.sum(X, axis = 1)
```

```python
# Build a simple neural network model
model = keras.Sequential([
    keras.Input(shape = (2,)),
	layers.Dense(4, activation = "relu"),
	layers.Dense(1)
])

# Print the model summary
model.summary()

# Compile the model
model.compile(optimizer = "adam", loss = "mean_squared_error")

# Train the model
model.fit(X, y, epochs=100, batch_size=32)

# Test the model
test_input = np.array([[42, 58]])  # Example input
predicted_sum = model.predict(test_input)
print(f"Predicted sum: {predicted_sum[0][0]}")
```
